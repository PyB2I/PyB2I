#  Encapsulamiento en Python

El **encapsulamiento** es un "principio" o "pilar" de la Programación Oriendata a Objetos que pretende mantener fuera del "alcance" o fuera del "ámbito" de acceso los atributos y también los métodos de una clase.

---

La filosofía en Python es que los programadores respeten las normas de codificación, de este modo se entiende que si una variable se declar a nivel de módulo es para que todos la puedan acceder.

Desde una función es posible acceder a variables declaradas fuera en un ámbito superior solo en modalidad de "lectura".

El ámbito de una variable declarada en una función es el cuerpo de la función, para modificar el valor de variables declaradas fuera de la función se debe utilizar la sentencia global.

En Python las "clásicas" definiciones sobre Encapsulamiento están sobreentendidas por los programdores mediante el siguiente acuerdo:

Declaraciones que comienzan con "\_" (guión bajo) se entiende que son **protected**.

Declaraciones que comienzan con "\_\_" (doble guión bajo) se entiende que son **private**.

En Programación Orientada a Objetos se asume que el ámbito de una declaración **protected** es donde se declara y toda su jerarquía de clases, mientras que el ámbito de una declaración **private** es solomanete donde se declara.

```python

class Person() :
    u""" Person digital definition """
    
    def __init__(self, first_name = "", last_name = "") :
        u""" Person constructor
        Initialize a Person objet
        """
        self.__first_name = first_name
        self.__last_name = last_name

        self.first_name = first_name

        return
    
    def setName(self, first_name) :
        u""" set first name value """
        self.__first_name = first_name
        
        return
    
    def __str__(self) :
        u""" String with person data """
        return self.__first_name + " " + self.__last_name + " [" + self.first_name +"]"
    
# ----


julio = Person("Julio", "Tentor")
print(julio)

julio.__first_name = "Julio Hernán"
julio.first_name = "Julio Hernán"
print(julio)

julio.setName("Julio Hernán")
print(julio)
```

## Property

Mediante el decorador **@property** Python facilita los mecanismos de getter/setter que existen en otros lenguajes de programación.

Esto incorpora una gran versatilidad desde el punto de vista del encapsulamiento. La primera versión de una clase puede implementar atributos, campos o variables de instancia y luego convertirlos en propiedades incorporando el código necesario para el getter y/o setter.

```python

class Person() :
    u""" Person digital definition """
    
    def __init__(self, first_name = "", last_name = "") :
        u""" Person constructor
        Initialize a Person objet
        """
        self.__first_name = first_name
        self.__last_name = last_name

        self.__history = ""
        return

    
    @property
    def first_name(self) :
        return self.__first_name

    @first_name.setter
    def first_name(self, value) :
        self.__history += "\nchange " + self.__first_name + " to " + value
        self.__first_name = value

        
    def __str__(self) :
        u""" String with person data """
        return self.__first_name + " " + self.__last_name

    
    def history(self) :
        return self.__history
        
# ----


julio = Person("Julio", "Tentor")
print(julio)

julio.__first_name = "Julio Hernán"
julio.first_name = "Julio Hernán"
print(julio)

print(julio.history())

```
